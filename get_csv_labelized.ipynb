Ne pas oublier d'ajouter la nouvelle page au concat total dans la cellule ci dessous

```python
full_df = pd.concat([page1, page2, page3, page4, page5], ignore_index=True)
print(full_df)
```

```python
full_df["bool_audible"] = [1 if i > 0 else 0 for i in full_df.mediane]
# db_labellized = full_df.drop('mediane', axis=1)
full_df = full_df.replace({'mediane':{-2:0, -1:1, 0:2, 1:3, 2:4}})
full_df
```

```python
full_df['filename'] = full_df['filename'].str.replace(r'-;$', '.wav')
full_df['filename'] = full_df['filename'].str.replace(r'-$', '.wav')
full_df['filename'] = full_df['filename'].map(lambda x: str(x)[1:])
# full_df.astype({'mediane': 'int32'}).dtypes
full_df['mediane'] = pd.to_numeric(full_df['mediane'], downcast='integer')
full_df
```
